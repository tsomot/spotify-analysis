## Unsupervised learning

Unsupervised learning models
- All of the songs in the dataset are songs that I enjoy.
- Unlike radio/preferences, dataset does not include songs I do not enjoy.
- Can't discriminate between songs I like and dislike, because dataset only contains songs I like.
- Not able to produce a supervised learning model. 
- Can try clustering and principal component analysis to see if any significant patterns emerge.

Methodology / prep:
- hold out on most recent playlist, November 2017
- assess model accuracy by looking at prediction probabilities for the tracks on this playlist resulting from each model

- train test split for each one
- random seed=42

Additional data cleaning: numerical values only

### n_components = 2, n_clusters=8

```python
# Visualize results on PCA-reduced data
reduced_data = PCA(n_components=2).fit_transform(train_scaled)
kmeans = KMeans()
kmeans.fit(reduced_data)
```

```python
# Plot

# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .02     # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on the dataset (PCA-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
